```python
import pandas as pd
import numpy as np

# Create synthetic weather dataset
data = {
    'Date': pd.date_range(start='2024-01-01', periods=20),
    'Temperature': np.random.normal(25, 5, 20),
    'Humidity': np.random.normal(60, 10, 20),
    'Rainfall': np.random.uniform(0, 10, 20),
    'WindSpeed': np.random.normal(10, 3, 20)
}
df = pd.DataFrame(data)

# Introduce missing values
df.loc[3, 'Temperature'] = np.nan
df.loc[5, 'Humidity'] = np.nan
df.loc[8, 'Rainfall'] = np.nan

# Introduce outlier
df.loc[10, 'Temperature'] = 50

print(df.head())
```

```python
import matplotlib.pyplot as plt

# Scatter Plot: Temperature vs Humidity
plt.figure(figsize=(8,6))
plt.scatter(df['Temperature'], df['Humidity'])
plt.title('Scatter Plot: Temperature vs Humidity')
plt.xlabel('Temperature')
plt.ylabel('Humidity')
plt.show()

# Histogram: Distribution of Temperature
plt.figure(figsize=(8,6))
plt.hist(df['Temperature'], bins=10)
plt.title('Histogram of Temperature')
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.show()
```

```python
# Before handling missing values
print("Before handling missing values:")
print(df)

# Detect missing
print("\nMissing values count:")
print(df.isnull().sum())

# Handle missing: fill with mean
df.fillna(df.mean(numeric_only=True), inplace=True)

print("\nAfter handling missing values:")
print(df)
```

```python
# Outlier detection using IQR for Temperature
print("Before outlier removal:")
print(df['Temperature'])

Q1 = df['Temperature'].quantile(0.25)
Q3 = df['Temperature'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("\nIQR Calculation:")
print(f"Q1: {Q1}, Q3: {Q3}, IQR: {IQR}")
print(f"Lower bound: {lower_bound}, Upper bound: {upper_bound}")

# Identify outliers
outliers = df[(df['Temperature'] < lower_bound) | (df['Temperature'] > upper_bound)]
print("\nIdentified outliers:")
print(outliers)

# Remove outliers
df = df[(df['Temperature'] >= lower_bound) & (df['Temperature'] <= upper_bound)]

print("\nAfter outlier removal:")
print(df)
```

```python
import pandas as pd
import numpy as np
# Create synthetic weather dataset
data = {
    'Date': pd.date_range(start='2024-01-01', periods=20),
    'Temperature': np.random.normal(25, 5, 20),
    'Humidity': np.random.normal(60, 10, 20),
    'Rainfall': np.random.uniform(0, 10, 20),
    'WindSpeed': np.random.normal(10, 3, 20)
}
df = pd.DataFrame(data)
```

```python
# Before scaling and discretization
print("Before:")
print(df)

# Scaling: Min-Max for Temperature
min_temp = df['Temperature'].min()
max_temp = df['Temperature'].max()
df['Temperature_scaled'] = (df['Temperature'] - min_temp) / (max_temp - min_temp)

# Discretization: Bin Temperature into categories
df['Temp_category'] = pd.cut(df['Temperature'], bins=3, labels=['Low', 'Medium', 'High'])

print("\nAfter:")
print(df)
```

```python
print("DataFrame Info:")
df.info()

print("\nDataFrame Describe:")
print(df.describe())
```

```python
print("\nCentral Tendency for Temperature:")
min_val = df['Temperature'].min()
max_val = df['Temperature'].max()
mean_val = df['Temperature'].mean()
median_val = df['Temperature'].median()
mode_val = df['Temperature'].mode()[0] if not df['Temperature'].mode().empty else 'No mode'

print(f"Min: {min_val}")
print(f"Max: {max_val}")
print(f"Mean: {mean_val}")
print(f"Median: {median_val}")
print(f"Mode: {mode_val}")
```

```python
print("\nDispersion for Temperature:")
range_val = max_val - min_val
q1 = df['Temperature'].quantile(0.25)
q3 = df['Temperature'].quantile(0.75)
iqr = q3 - q1
var = df['Temperature'].var()
std = df['Temperature'].std()

print(f"Range: {range_val}")
print(f"Q1: {q1}")
print(f"Q3: {q3}")
print(f"IQR: {iqr}")
print(f"Variance: {var}")
print(f"Standard Deviation: {std}")
```
